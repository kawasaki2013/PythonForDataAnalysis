### 複数のファイルを読み込んで一つのデータフレームにする

```python
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = './pydata-book/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    # データフレームの追加をして、データフレームが要素のリストオブジェクトを作る
    pieces.append(frame)
    
```

```python
#pieces
#frame
```

```python
# piecesに格納されたそれぞれの要素を１つのデータフレームオブジェクトにまとめる
names = pd.concat(pieces, ignore_index=True)
```

* pandas.concat → 行方向に追加する
* ignore_index=True → オリジナルのindexを消す

```python
names
```

* 巨大なデータフレームを作ったので、あとはこれまでのテクニックで分析していく

```python
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
total_births.tail()
```

```python
total_births.plot(title='Total births by sex and year')
```

### 名前の出現割合を求める

* まずは、年代別・性別の出生数割合を求める
    * propという関数を定義する

```python
# 年代別・性別の出生数割合を求める関数
def add_prop(group):
    # 整数型の割り算だと0になっちゃうので、float型にキャストする
    births = group.births.astype(float)
    # groupオブジェクトにprop列を加える
    ## 分子には出生数。分母には年代と性毎の出生数の総和
    group['prop'] = births / births.sum()
    return group
```

```python
# yearとsexでグループ化して、そのグループのそれぞれにadd_prop関数を適用する
names = names.groupby(['year', 'sex']).apply(add_prop)
names[:5]
```
